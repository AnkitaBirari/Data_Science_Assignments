```python
# import dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#load the dataset
book = pd.read_csv("/content/book.csv")
book
```

```python
# EDA process
book.info()
```

```python
book.describe()
```

```python
book.shape
```

## Association rules with 10% support ans 70% confidence

```python
frequent_itemsets = apriori(book, min_support=0.1, use_colnames = True) # Apriori  algorithm
frequent_itemsets
```

```python
# with 70% confidence 
rules = association_rules(frequent_itemsets, metric = "lift",min_threshold=0.9)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:10]
```

```python
rules[rules.lift>1]
```

```python
# Visualiztion of obtaioned rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
