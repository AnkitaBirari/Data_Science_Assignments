```python
# converting Long data into wide data using pivote table
book2 = book1.pivot_table(index="User.ID",columns = "Book.Title",values = "Book.Rating").reset_index(drop=True)
book2
```

```python
book2.index=book1['User.ID'].unique()
book2
```

```python
book2.fillna(0,inplace = True)
book2
```

```python
user_sim = 1-pairwise_distances(book2.values,metric = 'cosine')
user_sim
```

```python
user_sim2 = pd.DataFrame(user_sim)
user_sim2
```

```python
user_sim2.index = book1['User.ID'].unique()
user_sim2.columns = book1['User.ID'].unique()
user_sim2
```

```python
np.fill_diagonal(user_sim,0)
```
