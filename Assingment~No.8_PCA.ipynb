## PCA Implementation

```python
pca = PCA()
pca_values = pca.fit_transform(wine3)
pca_values
```

```python
# PCA Componenrs matriz or covarience matrix
pca.components_
```

```python
# vaience 
variance = pca.explained_variance_ratio_
variance
```

```python
#Cummulative varience of each PCA
var1 = np.cumsum(np.round(variance,4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.figure(figsize=(6,4))
plt.plot(var1,color = 'purple')
plt.show()
```

```python
# Final DataFrame
final_df = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis = 1)
final_df
```

```python
# Visualization of  PCAs
fig = plt.figure(figsize=(12,8))
sns.scatterplot(data=final_df)
plt.show()
```

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we have already normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine3,'complete'))
plt.show()
```

```python
#Create cluster (y)
hclusters = AgglomerativeClustering(n_clusters =3, affinity = 'euclidean',linkage = 'ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine3),columns = ['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding the column clustersid to dataset
wine4 = wine.copy()
wine4['clustersid'] = hclusters.labels_
wine4
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range (1,6):
  kmeans = KMeans(n_clusters=i,random_state=2)
  kmeans.fit(wine3)
  wcss.append(kmeans.inertia_)
```

```python
# plot k values range vs WCSS to get Elbow graph for choosing k(no.of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel("WCSS")
plt.show()
```

```python
# cluster algorithm using k = 3
clusters3 = KMeans(3,random_state=30).fit(wine3)
clusters3
```

```python
clusters3.labels_
```

```python
# Adding the clustersid column to the dataset
wine4 = wine.copy()
wine4['clustersid']=clusters3.labels_
wine4
```
